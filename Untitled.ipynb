```python
#PARTE2: Con esa tabla de valores procederemos a estimar la Evapotranspiración ET0 mensual promedio.
#utilizando las 3 ecuaciones siguientes 
#1.- Ec 7, Jansen & Haise (1963)->ETo= (0.0252T+0.078)Rs 
#2.-Ec. 31, Valiantzas 1 (2012):  𝜑 es la latitud en radianes ->
#-> ETo= 0.0393 RS (Tmean+9.5)^(0.5)-(0.19)Rs^(0.6)phi^(0.15)+0.0061(Tmean+20)(1.12Tmean-Tmin-2)^(0.7)
#3.-Ec. 32,  Valiantzas 4 (2013): α es el albedo (Albedo_Avg),
#Ra es la radiación solar en la parte alta de la atmósfera (ver ecuación 21, página 46 del Reporte FAO 56),
#u2 es la velocidad del viento a 2 m de altura (WS_ms_S_WVT), z es la altura sobre el nivel mar.
```

```python
#Tenemos nuestra tabla de valores
dft.head()
```

```python
#Veamos la primer ecuación 
#1.- Ec 7, Jansen & Haise (1963)->ETo= (0.0252T+0.078)Rs
jh=[]
#Usemos la ecuación dada

for i in range (0,len(dft)):
    j= ((0.0252)*(dfm.AirTC_Avg[i]) + 0.078)*(dfc.Rs[i])
    jh.append(j)
    
#Metamos el valor obtenido en una columna. 
dft["EvoTJH"]=jh
dft.head()
#dft["EvoTJH"].head()
```

```python
dfm.AirTC_Avg[0]
```

```python
#Veamos la segunda ecuación
#2-Ec. 31, Valiantzas 1 (2012):  𝜑 es la latitud en radianes ->
#-> ETo= 0.0393 RS (Tmean+9.5)^(0.5)-(0.19)Rs^(0.6)phi^(0.15)+0.0061(Tmean+20)(1.12Tmean-Tmin-2)^(0.7)
```

```python
Val1=[]
#Usemos la ecuación dada
for i in range (0,len(dft)):
    #Para no hacernos bolas con los paréntesis
    phi=(dft.Latitud[i])*(math.pi/180)
    tp=float(dft.Tp[i])
    tm=dft.Tmin[i]
    V1=(0.0393*dft.Rs[i]*(tp+9.5)**(0.5))-((0.19*(dft.Rs[i])**(0.6))*phi**(0.15))+(0.0061*(tp+20)*(1.12*tp-tm-2)**(0.7))
    
    Val1.append(V1)

#Metamos el valor obtenido en una columna
dft["EvoTVali1"]=Val1

dft.head()
```

```python
#3.-Ec. 32,  Valiantzas 4 (2013): α es el albedo (Albedo_Avg),
#-> 0.051(1-α)Rs(Tmean+9.5)^(0.5)-2.4(Rs/Ra)^2+0.048(Tmean+20)(1-RH/100)(0.5+0.536u2)+0.00012z
#Ra es la radiación solar en la parte alta de la atmósfera (ver ecuación 21, página 46 del Reporte FAO 56),
#u2 es la velocidad del viento a 2 m de altura (WS_ms_S_WVT), z es la altura sobre el nivel mar.
```

```python
Val4=[]
#Usemos la ecuación dada
for i in range (0,len(dft)):
    #Para no hacernos bolas con los paréntesis
    al=dfm.albedo_Avg[i]
    tp=dfc.Tp[i]
    tm=dft.Tmin[i]
    RH=dfc.RH[i]
    #La constante solar
    G=0.0820 
    #Sea J el número de día en el año entre 1 y 365 (entre 1ene y 31dic). Por lo que
    if i==0:
        J=dfc.Día[0]
    elif dfc.Día[i] != dfc.Día[i-1]:
        J=J+1
    elif dfc.Día[i] == dfc.Día[i-1]:
        J=J
    #La distancia relativa tierra sol
    dr=1+0.033*math.cos(2*3.1416*J/365)
    #Declinación solar 
    d=0.409*math.sin((2*3.1416*J/365)-1.39)
    #El ángulo de la hora de la puesta de sol
    w=math.acos(-math.tan(phi) * math.tan(d))
    #sea u la velocidad del viento a 2 metros de altura
    u=dft.Vp[i]
    #Sea z la elevación sobre el nivel del mar
    z=dft.Elevación[i]
    
    
    #calculando la radiación atmosfera
    Ra=(24*60*G*dr/3.1416)*(w*math.sin(phi)*math.sin(d)+math.cos(phi)*math.cos(d)*math.sin(w))
    
    V4=(0.051*(1-al)*dft.Rs[i]*(tp+9.5)**(0.5))-(2.4*(dfc.Rs[i]/Ra)**2)+0.048*(tp+20)*(1-RH/100)*(0.5+0.536*u)+0.00012*z
    
    Val4.append(V4)

#Metamos el valor obtenido en una columna
dft["EvoTVali4"]=Val4

dft.head()
```
