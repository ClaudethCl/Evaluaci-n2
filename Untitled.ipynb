```python
#librerías 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import datetime
```

```python
#En la planeación de irrigación de cultivos o estudios de uso de agua en la Agricultura, 
#se requiere conocer la cantidad de vapor de agua en la atmósfera que proviene por un lado
#de la evaporación de la humedad de suelo, así como también de la transpiración de las plantas. 
#A este proceso se le conoce como Evapotranspiración.
#La Organización de las Naciones Unidas para la Alimentación y la Agricultura (FAO), publicó su reporte número 56   
#(o en HTML), para calcular la Evapotranspiración y estimar el uso de agua en los cultivos. 
#El método que utilizó la FAO para modelar la Evapotranspiración fue la ecuación de Penman-Monteith. 
#Esta ecuación requiere conocer la temperatura diaria promedio, la velocidad del viento,
#la humedad relativa y la radiación solar.  
#La Evapotranspiración de referencia ET0, es uno de los parámetros mas importantes en los estudios hidrológicos, 
#ambientales y agrícolas y juega un papel muy importante en los proyectos de manejo de irrigación
#y uso de agua en la agricultura. La ET0 es estimada por diversos métodos: utilizando lisímetros, 
#sistemas de covariancias turbulentas o utilizando métodos indirectos utilizando variables climáticas. 
#El requerimiento de conocer un conjunto de valores de las variables climáticas ha sido una de las limitantes
#de la aplicación de la Ecuación de Penman-Monteith. Por ello se han desarrollado toda una serie de ecuaciones
#para el cálculo alternativo de la ET0 bajo diversas condiciones climáticas. En esta actividad nos daremos
#la tarea de evaluar cuáles son las mejores alternativas de la ecuación de Penman-Monteith
#para una región climática como la nuestra (zona semiárida seca).
```

```python
#PARTE2: Con esa tabla de valores procederemos a estimar la Evapotranspiración ET0 mensual promedio.
#utilizando las 3 ecuaciones siguientes 
#1.- Ec 7, Jansen & Haise (1963)->ETo= (0.0252T+0.078)Rs 
#2.-Ec. 31, Valiantzas 1 (2012):  𝜑 es la latitud en radianes ->
#-> ETo= 0.0393 RS (Tmean+9.5)^(0.5)-(0.19)Rs^(0.6)phi^(0.15)+0.0061(Tmean+20)(1.12Tmean-Tmin-2)^(0.7)
#3.-Ec. 32,  Valiantzas 4 (2013): α es el albedo (Albedo_Avg),
#Ra es la radiación solar en la parte alta de la atmósfera (ver ecuación 21, página 46 del Reporte FAO 56),
#u2 es la velocidad del viento a 2 m de altura (WS_ms_S_WVT), z es la altura sobre el nivel mar.
```

```python
#Tenemos nuestra tabla de valores
df1.head()
```

```python
#Veamos la primer ecuación 
#1.- Ec 7, Jansen & Haise (1963)->ETo= (0.0252T+0.078)Rs
jh=[]
#Usemos la ecuación dada

for i in range (0,len(df1)):
    j= (0.0252*float(dfm.TempAire[i]) + 0.078)*df1.Rs[i]
    jh.append(j)
    
#Metamos el valor obtenido en una columna. 
df1["EvoTJH"]=jh
df1.head()
```

```python
#Veamos la segunda ecuación
#2-Ec. 31, Valiantzas 1 (2012):  𝜑 es la latitud en radianes ->
#-> ETo= 0.0393 RS (Tmean+9.5)^(0.5)-(0.19)Rs^(0.6)phi^(0.15)+0.0061(Tmean+20)(1.12Tmean-Tmin-2)^(0.7)
```

```python
Val1=[]
#Usemos la ecuación dada
for i in range (0,len(df1)):
    #Para no hacernos bolas con los paréntesis
    phi=(df1.Latitud[i])*(3.1416/180)
    tp=float(dfm.Tmean[i])
    tm=df1.Tmin[i]
    V1=(0.0393*df1.Rs[i]*(tp+9.5)**(0.5))-((0.19*(df1.Rs[i])**(0.6))*phi**(0.15))+(0.0061*(tp+20)*(1.12*tp-tm-2)**(0.7))
    
    Val1.append(V1)

#Metamos el valor obtenido en una columna
df1["EvoTVali1"]=Val1

df1.head()
```

```python
#3.-Ec. 32,  Valiantzas 4 (2013): α es el albedo (Albedo_Avg),
#-> 0.051(1-α)Rs(Tmean+9.5)^(0.5)-2.4(Rs/Ra)^2+0.048(Tmean+20)(1-RH/100)(0.5+0.536u2)+0.00012z
#Ra es la radiación solar en la parte alta de la atmósfera (ver ecuación 21, página 46 del Reporte FAO 56),
#u2 es la velocidad del viento a 2 m de altura (WS_ms_S_WVT), z es la altura sobre el nivel mar.
```

```python
Val4=[]
#Usemos la ecuación dada
for i in range (0,len(df1)):
    #Para no hacernos bolas con los paréntesis
    al=float(dfm.albedo_Avg[i])
    tp=float(dfm.Tmean[i])
    tm=df1.Tmin[i]
    RH=df1.HumRel[i]
    #La constante solar
    G=0.0820 
    #Sea J el número de día en el año entre 1 y 365 (entre 1ene y 31dic). Por lo que
    if i==0:
        J=dfm.Dia[0]
    elif dfm.Dia[i] != dfm.Dia[i-1]:
        J=J+1
    elif dfm.Dia[i] == dfm.Dia[i-1]:
        J=J
    #La distancia relativa tierra sol
    dr=1+0.033*math.cos(2*3.1416*J/365)
    #Declinación solar 
    d=0.409*math.sin((2*3.1416*J/365)-1.39)
    #El ángulo de la hora de la puesta de sol
    w=math.acos(-math.tan(phi) * math.tan(d))
    #sea u la velocidad del viento a 2 metros de altura
    u=df1.VelVientProm[i]
    #Sea z la elevación sobre el nivel del mar
    z=df1.Elevación[i]
    
    
    #calculando la radiación atmosfera
    Ra=(24*60*G*dr/3.1416)*(w*math.sin(phi)*math.sin(d)+math.cos(phi)*math.cos(d)*math.sin(w))
    
    V4=(0.051*(1-al)*df1.Rs[i]*(tp+9.5)**(0.5))-(2.4*(df1.Rs[i]/Ra)**2)+0.048*(tp+20)*(1-RH/100)*(0.5+0.536*u)+0.00012*z
    
    Val4.append(V4)

#Metamos el valor obtenido en una columna
df1["EvoTVali4"]=Val4

df1.head()
```
